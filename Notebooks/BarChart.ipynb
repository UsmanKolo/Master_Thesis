```python
import numpy as np
import matplotlib.pyplot as plt

folds = ['4', '5', '6', '7', '8', '9', '10', '11']
RMSE = np.array([7.783, 7.563, 8.071, 9.161, 9.524, 8.792, 8.182, 7.983])
MSE =  np.array([60.578, 57.2, 65.148, 83.918, 90.705, 77.307, 66.947, 63.722])
MAE =  np.array([6.216, 6.1, 6.489, 7.316, 7.684, 7.094, 6.661, 6.662])
R2  =  np.array([56.528, 58.724, 54.2, 39.677, 28.723, 35.002, 40.319, 39.681])

ind = [x for x, _ in enumerate(folds)]

plt.figure(figsize=(10,10))
plt.bar(ind, R2, width=0.8, label='R2',bottom=MAE+MSE+RMSE)
plt.bar(ind, MAE, width=0.8, label='MAE', bottom=MSE+RMSE)
plt.bar(ind, MSE, width=0.8, label='MSE', bottom=RMSE)
plt.bar(ind, RMSE, width=0.8, label='RMSE')

plt.xlabel('Years', fontweight='bold', fontsize=13)
plt.ylabel('Scores', fontweight='bold', fontsize=13)
plt.title('Metrics for 2-Layered Simple LSTM with 50 Neurons', fontweight='bold', fontsize=13)

plt.xticks(ind, folds)
plt.tick_params(labelsize=13)
plt.legend(loc='best')
plt.show()
```

```python
# set width of bars
barWidth = 0.2
folds = ['4', '5', '6', '7', '8', '9', '10', '11']

# set heights of bars
RMSE = [7.783, 7.563, 8.071, 9.161, 9.524, 8.792, 8.182, 7.983]
MSE = [60.578, 57.2, 65.148, 83.918, 90.705, 77.307, 66.947, 63.722]
MAE = [6.216, 6.1, 6.489, 7.316, 7.684, 7.094, 6.661, 6.662]
R2 = [56.528, 58.724, 54.2, 39.677, 28.723, 35.002, 40.319, 39.681]
 
# Set position of bar on X axis
r1 = np.arange(len(RMSE))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.figure(figsize=(11,11))

# Make the plot
plt.bar(r1, RMSE, width=barWidth, edgecolor='white', label='RMSE')
plt.bar(r2, MSE, width=barWidth, edgecolor='white', label='MSE')
plt.bar(r3, MAE, width=barWidth, edgecolor='white', label='MAE')
plt.bar(r4, R2, width=barWidth, edgecolor='white', label='MAE')
 
# Add xticks on the middle of the group bars
plt.xlabel('Folds', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(RMSE))], folds)
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
