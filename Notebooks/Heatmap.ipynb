```python
data = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\QBElecWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data1 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\QBWaterWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data2 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\QBGasWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data3 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\HUElecWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data4 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\HUWaterWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data5 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\HUGasWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data6 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\GTWYElecWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data7 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\GTWYWaterWeatherDailyData.csv',
 index_col=0, parse_dates=True)

data8 = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\GTWYGasWeatherDailyData.csv',
 index_col=0, parse_dates=True)
```

```python
x = data.values
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled, columns = data.columns)
df.describe()
```

```python
plt.figure(figsize=(10,10))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
sns.set(font_scale=2)
sns.heatmap(corr, mask=mask, center=0,square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
plt.xticks(rotation='vertical')
plt.yticks(rotation='horizontal')
plt.xlabel('Features',fontsize=20, ha='center')
plt.ylabel('Features',fontsize=20, ha='center')
plt.title('Heatmap Between Features Queens Building Electric Dataset',fontsize=20, ha='center')
plt.tick_params(labelsize=16)
plt.ylim(6, 0)
plt.tight_layout()
```
