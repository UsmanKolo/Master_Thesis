```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,9)
plt.style.use('bmh')
import numpy as np
from sklearn.ensemble import *
import seaborn as sns
from sklearn.preprocessing import *
```

```python
GTWYHourly = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\GTWYHourly.csv', index_col=0, parse_dates=True)
GTWYDaily = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\GTWYDaily.csv', index_col=0, parse_dates=True)
HUHourly = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\HUHourly.csv', index_col=0, parse_dates=True)
HUDaily = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\HUDaily.csv', index_col=0, parse_dates=True)
QBHourly = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\QBHourly.csv', index_col=0, parse_dates=True)
QBDaily = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\QBDaily.csv', index_col=0, parse_dates=True)
```

```python
x = GTWYHourly.values
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled, columns = GTWYHourly.columns)
df.describe()
```

```python
plt.figure(figsize=(10,10))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
sns.set(font_scale=1)
sns.heatmap(corr, mask=mask, center=0,square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
plt.xticks(rotation='vertical')
plt.yticks(rotation='horizontal')
plt.xlabel('Features',fontsize=20, ha='center')
plt.ylabel('Features',fontsize=20, ha='center')
plt.title('Heatmap Between Features Gateway House Hourly Dataset',fontsize=20, ha='center')
plt.tick_params(labelsize=16)
plt.ylim(8, 0)
plt.tight_layout()
```
