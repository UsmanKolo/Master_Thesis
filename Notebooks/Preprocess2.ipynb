```python
# Weather data
weather = pd.read_csv('C:\\Users\\moham\\OneDrive - De Montfort University\\Final Project\\Master_Thesis\\Dataset\\Weather.csv')
weather.head()
```

```python
weather = weather[['dt_iso', 'temp', 'dew_point', 'pressure', 'humidity', 'wind_speed']]
weather = weather.rename(columns={'dt_iso': 'date'})
```

```python
weather['date'] = pd.to_datetime(weather['date'])
weather.head()
```

```python
newdf = weather.copy()
newdf = newdf.set_index('date')
newdf.head()
```

```python
hourly_weather = newdf.copy()
daily_weather = newdf.resample(rule='d').mean().copy()
```
